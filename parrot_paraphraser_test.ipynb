```python
from parrot import Parrot

parrot = Parrot()
```

```python
phrases = [
  "Which all places are worth to visit in Pune?"
]

for phrase in phrases:
  print("-"*100)
  print("Input_phrase: ", phrase)
  print("-"*100)
  paraphrases = parrot.augment(input_phrase=phrase)
  for paraphrase in paraphrases:
   print(paraphrase)
```
